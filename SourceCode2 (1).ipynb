```python
import torch
import pandas as pd
```

```python
correct_sentences = []
with open('./data/train.txt', 'r', encoding='utf-8') as file:
    cnt = 0
    for line in file:
        fields = line.strip().split('\t')
        if len(fields) > 2:
            print(f"error when parsing line{cnt}")
        else:
            correct = fields[0].strip()
            correct_sentences.append(correct)
```

```python
import random
import string
```

```python
original_sentence = "Jeremy returns to London to look for Diana."
```

```python
def add_typo(sentence):
    if len(sentence) > 1:
        sentence = list(sentence)
        index = random.randint(0, len(sentence) - 1)
        original_char = sentence[index]

        if original_char in string.ascii_letters:
            if original_char.islower():
                possible_chars = [char for char in string.ascii_letters if char != original_char and char.islower()]
            else:
                possible_chars = [char for char in string.ascii_letters if char != original_char and char.isupper()]
        else:
            possible_chars = [char for char in string.ascii_letters if char != original_char]

        new_char = random.choice(possible_chars)
        sentence[index] = new_char
        return ''.join(sentence)

    return sentence
```

```python
add_typo(original_sentence)
```

```python
def delete_word(sentence):
    words = sentence.split()
    if len(words) > 1:
        index = random.randint(0, len(words) - 1)
        del words[index]
        return ' '.join(words)
    return sentence
```

```python
delete_word(original_sentence)
```

```python
def switch_adjacent_words(sentence):
    words = sentence.split()
    if len(words) > 1:
        index = random.randint(0, len(words) - 2)
        words[index], words[index + 1] = words[index + 1], words[index]
        return ' '.join(words)
    return sentence
```

```python
switch_adjacent_words(original_sentence)
```

```python
def replace_random_word(sentence):
    words = sentence.split()
    index_to_replace = random.randint(0, len(words) - 1)
    word_to_replace = words[index_to_replace]
    random_sentence = random.choice(correct_sentences)
    replacement_words = random_sentence.split()
    replacement_word = random.choice([word for word in replacement_words if word != word_to_replace])
    words[index_to_replace] = replacement_word
    sentence = ' '.join(words)
    return sentence
```

```python
replace_random_word(original_sentence)
```

```python
def corrupt_sentence(sentence):
    action = random.choice(['typo', 'delete', 'swap', 'replace'])

    if action == 'typo':
        return add_typo(sentence)
    elif action == 'delete':
        return delete_word(sentence)
    elif action == 'swap':
        return switch_adjacent_words(sentence)
    elif action == 'replace':
        return replace_random_word(sentence)

```

```python
corrupt_sentence(original_sentence)
```

```python
output_file_path = './part2.txt'

with open(output_file_path, 'w') as file:
    for sentence in correct_sentences:
        corrupted_sentence = corrupt_sentence(sentence)
        file.write(f"{sentence}\t{corrupted_sentence}\n")
```
